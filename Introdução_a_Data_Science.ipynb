```python
movies= pd.read_csv('tmdb_5000_movies.csv')
```

```python
media= movies.runtime.mean()
print(media/60)
```

```python
 #Quando extraimos só uma coluna chamamos de série, neste caso uma única série
 movies.original_language.value_counts()
```

```python
 movies.original_language.value_counts().index
```

```python
 movies.original_language.value_counts().values
```

```python
 #criando um dataframe, neste caso temos um índice(lingua) e uma coluna(valores numéricos)
 movies.original_language.value_counts().to_frame()
```

```python
 #neste caso o índice vira um contador
 contagem_de_lingua= movies.original_language.value_counts().to_frame().reset_index()
 contagem_de_lingua.columns= ['original_language', 'total']
 contagem_de_lingua.head()
```

```python
sns.barplot(x='original_language', y='total', data=contagem_de_lingua)
```

```python
#utilizando a função catplot() que gera gráficos por categorias
sns.catplot(x="original_language", kind="count", data= movies)
```

```python
plt.pie(contagem_de_lingua['total'], labels= contagem_de_lingua['original_language'])
```

```python
total_por_lingua= movies['original_language'].value_counts()
total_geral= total_por_lingua.sum()
total_de_ingles= total_por_lingua.loc["en"]
total_do_resto= total_geral - total_de_ingles
print(total_de_ingles, total_do_resto)
```

```python
dados= {'lingua': ['ingles', 'outros'],
      'total': [total_de_ingles, total_do_resto]
}

```

```python
pd.DataFrame(dados)
```

```python
sns.barplot(x= 'lingua', y='total', data= dados)
```

```python
plt.pie(dados['total'], labels= dados['lingua'])
```

```python
total_por_lingua_de_outros_filmes= movies.query("original_language != 'en'").original_language.value_counts()
total_por_lingua_de_outros_filmes
```

```python
#query são perguntas
filmes_sem_lingua_original_em_ingles= movies.query("original_language != 'en'")

#funções de alto nível são geradas diretamente sem ter como manipular certas coisas na unha, já funções de baixo
#nível dão está possibilidade
#aspect serve para mudar o tamanho do gráfico, order serve para decidir a ordem em que os dados serão apresentados
sns.catplot(x='original_language', kind= 'count', data= filmes_sem_lingua_original_em_ingles, aspect=2,
            palette="GnBu_d",
            order= total_por_lingua_de_outros_filmes.index,)
```
